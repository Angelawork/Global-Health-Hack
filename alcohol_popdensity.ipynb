```python
import pandas as pd
import numpy as np
```

```python
den = pd.read_csv('density.csv')
```

```python
den.head()
```

```python
den.columns
```

```python
cols = ['Country Code','Indicator Name','Indicator Code','Unnamed: 67']
den = den.drop(columns=cols)
den.head()
```

```python
df = den.melt(id_vars=['Country Name'], var_name='Year', value_name='Population Density')
df.head()
```

```python
df['Year'] = df['Year'].astype(int)
df = df.drop(df[(df['Year'] > 2019) | (df['Year'] < 2015)].index)
df.head()
```

```python
df = df.rename(columns={'Country Name':'Country'})
```

```python
alcohol = pd.read_csv('alcohol.csv')
alcohol.head()
```

```python
alcohol.shape
```

```python
alcohol.columns
```

```python
alcohol = alcohol.rename(columns = {'Prevalence - Alcohol use disorders - Sex: Both - Age: Age-standardized (Percent)':'Alcohol Use Disorders Prevalence'})
```

```python
alcohol = alcohol.rename(columns = {'Entity':'Country'})
```

```python
alcohol = alcohol.drop('Code',axis=1)
```

```python
merged = pd.merge(df, alcohol, on=['Country', 'Year'])
merged.head()
```
